## 讀取Wafer SQC csv File

```python
import pandas as pd
import numpy as np
wafer_id = 'UJ1900816'
in_file = './WMA/WMA_V2/SQC/playnitride/PN_' + wafer_id + '_20211223_v0.csv'
parta_df = pd.read_csv(in_file, nrows=25, skiprows=1, usecols=[0,1], names=['name','value'])
parta_df['partb'] = np.nan
parta_df['partb_top3_min'] = ''
parta_df['partb_top3_max'] = ''
parta_df.set_index('name', inplace=True)
partb_df = pd.read_csv(in_file, skiprows=28)
if (partb_df['AREA'][-1:].values[0] == 'B0') and (partb_df['UPL_CHIP_PL'][-1:].isnull().values[0]) :
    print('0000000000000000')
    partb_df.drop([partb_df.shape[0]-1], inplace=True)
```

```python
parta_df['partb'].loc['LED_CNT_X'] = len([x for x in sorted(partb_df['COORDINATE_X'].astype(str).unique()) if str(x) != 'nan'])
parta_df['partb'].loc['LED_CNT_Y'] = len([x for x in sorted(partb_df['COORDINATE_Y'].astype(str).unique()) if str(x) != 'nan'])
parta_df['partb'].loc['LED_CNT_SUM'] = partb_df.shape[0]
parta_df['partb_top3_min'].loc['LED_CNT_X'] = sorted(partb_df['COORDINATE_X'].astype(str).unique())[:3]
parta_df['partb_top3_max'].loc['LED_CNT_X'] = sorted(partb_df['COORDINATE_X'].astype(str).unique())[-3:]
parta_df['partb_top3_min'].loc['LED_CNT_Y'] = sorted(partb_df['COORDINATE_Y'].astype(str).unique())[:3]
parta_df['partb_top3_max'].loc['LED_CNT_Y'] = sorted(partb_df['COORDINATE_Y'].astype(str).unique())[-3:]

parta_df['partb'].loc['LED_SHOT_SUM'] = len(partb_df['AREA'].unique())
parta_df['partb_top3_min'].loc['LED_SHOT_SUM'] = sorted(partb_df['AREA'].astype(str).unique())[:3]
parta_df['partb_top3_max'].loc['LED_SHOT_SUM'] = sorted(partb_df['AREA'].astype(str).unique())[-3:]

for ins_column in partb_df.columns :
    parta_df.loc[ins_column] = ['', len(partb_df[ins_column].unique()), sorted(partb_df[ins_column].astype(str).unique())[:3], sorted(partb_df[ins_column].astype(str).unique())[-3:]]

parta_df.to_csv(wafer_id + 'parta_df.csv')


print('LED_CNT_SUM - (LED_CNT_X * LED_CNT_X) = ', int(parta_df['value'].loc['LED_CNT_SUM']) - (int(parta_df['value'].loc['LED_CNT_X']) * int(parta_df['value'].loc['LED_CNT_X'])))



print('Check_LED_SUM partaX*Y = ', (int(parta_df['value'].loc['LED_CNT_X']) * int(parta_df['value'].loc['LED_CNT_Y'])))
print('Check_LED_SUM parta_sum = ', int(parta_df['value'].loc['LED_CNT_SUM']))
print('Check_LED_SUM partb_cnt = ', partb_df.shape[0])

print('Check_LED_DIM_X cnt, diff = ', int(parta_df['value'].loc['LED_CNT_X']), int(parta_df.loc['LED_CNT_X']['value']) - len(partb_df[partb_df['COORDINATE_X'].notnull()]['COORDINATE_X'].unique()))

print('Check_LED_DIM_Y cnt, diff = ', int(parta_df['value'].loc['LED_CNT_Y']), int(parta_df.loc['LED_CNT_Y']['value']) - len(partb_df[partb_df['COORDINATE_Y'].notnull()]['COORDINATE_Y'].unique()))


print(int(parta_df.loc['LED_CNT_X']['value']), int(parta_df.loc['LED_CNT_Y']['value']))

print(np.shape(partb_df)[0] - (int(parta_df['value'].loc['LED_CNT_X']) * int(parta_df['value'].loc['LED_CNT_X'])))

print(sorted(partb_df['UPL_CHIP_PL'].unique()))

x_list = sorted(partb_df['COORDINATE_X'].unique())
x_clearn_list = [x for x in x_list if str(x) != 'nan']
y_list = sorted(partb_df['COORDINATE_Y'].unique())
y_clearn_list = [x for x in y_list if str(x) != 'nan']
print('x_len, y_len = ', len(x_clearn_list), len(y_clearn_list))
```

## 讀取Wafer SQC cvs檔, 將指定檢查項目存成csv

```python
import pandas as pd
import numpy as np
wafer_list = ['UJ1900181', 'UJ1701426', 'UJ1501143']
for wafer_id in wafer_list : 
    in_file = './SQC/playnitride/new/PN_' + wafer_id + '_20211014_v0.csv'
    parta_df = pd.read_csv(in_file, nrows=25, skiprows=1, usecols=[0,1], names=['name','value'])
    parta_df.set_index('name', inplace=True)
    partb_df = pd.read_csv(in_file, skiprows=28)
    partb_df[partb_df['EL_VF_1UA']>0].to_csv('./SQC/playnitride/new/PN_' + wafer_id + '_byEL.csv')
    partb_df[partb_df['EPI_WP']>0].to_csv('./SQC/playnitride/new/PN_' + wafer_id + '_byEPI.csv')
```

## 讀取Wafer SQC cvs檔, 修改內容,  並另存成csv

```python
file_path = "./SQC/playnitride/PN_"
source_name = "UJ1501143"
targe_name = "UJ1501144"

import csv
file = open(file_path + source_name + "_1.csv", 'r')
csvCursor = csv.reader(file)
personDF = []
for row in csvCursor:
    personDF.append(row)
file.close()
csvHeader0 = [item.replace(source_name, targe_name) for item in personDF[5]]
personDF[5] = csvHeader0
file = open(file_path + targe_name + "_1.csv", 'w', newline='')
csvCursor = csv.writer(file)
csvCursor.writerows(personDF)
file.close()
```
